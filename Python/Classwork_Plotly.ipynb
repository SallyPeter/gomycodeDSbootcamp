```python
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv("used cars.csv")
```

```python
fig = px.bar(df, x="Brand", y="Cars Listings", color="Cars Listings", height=400)
```

```python
fig.update_layout(title_text="Cars Listings by Brand", xaxis_title="Brand", yaxis_title="Cars Listings")
```

```python
fig.update_traces(marker=dict(color='red', opacity=0.7, line=dict(color='red', width=1.5)))
fig.update_layout(coloraxis_colorbar=dict(title="Cars Listings"))
```

```python
fig = px.pie(df, values='Cars Listings', names='Brand',
             color='Brand', title='Engine Fuel Type Distribution of Cars')
```

```python
fig.update_traces(textposition='inside', textinfo='percent+label')
```

```python
fig.update_layout(uniformtext_minsize=14, uniformtext_mode='hide')
```

## **Instructions :**
This dataset contains the prevalence and trends of tobacco use for 1995-2010. Percentages are weighted to population characteristics.

1. Load the data into a Pandas DataFrame
2. Inspect your dataframe
3. Create a line chart to visualize the 'daily-smokers' evolution over time
4. Create a stacked area chart to compare the 'non-somkers' to 'former-smokers' rates within the same period

```python
data = pd.read_csv('tobacco.csv')

data.head()
```

```python
# Create a line chart to visualize the 'daily-smokers' evolution over time
fig = px.line(data, x='Year', y='Smoke everyday', title='Daily Smokers\' evolution over time')
fig.show()
```

```python
# Create a stacked area chart to compare the 'non-somkers' to 'former-smokers' rates within the same period
fig = px.area(data, x='Year', y=['Former smoker',	'Never smoked'], title='Former smokers vs Never Smoked')
fig.show()
```
