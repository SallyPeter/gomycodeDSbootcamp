*Make sure to import pandas library before doing the exercises.*

```python
import pandas as pd
```

```python
# Create a list of products, 'Apple', 'Banana', 'Carrot', 'Daikon'

products = ['Apple', 'Banana', 'Carrot', 'Daikon']
```

```python
# Convert the list into a pandas Series using the pd.Series() function
products = pd.Series(products, name='products')
```

```python
# Print out the new Series to see the values
products
```

```python
# Create a Series with prices of the products, [2, 1, 3, 4]
prices = [2, 1, 3, 4]
prices = pd.Series(prices, name='prices')
```

```python
# Print out the new Series to see the values
prices
```

```python
# Add the two Series together and print out the result.

product_and_prices = pd.concat([products, prices], axis = 1) # Concat defaults to row-wise concatenation while axis=1 changes this to column wise

product_and_prices
```

```python
# Create a DataFrame using a dictionary of lists, where the keys are 'Product' and 'Price' and the values are the two Series you created.
products = ['Apple', 'Banana', 'Carrot', 'Daikon']
prices = [2, 1, 3, 4]

product_and_prices_dict = {'Product':products, 'Price':prices}

new_product_and_prices = pd.DataFrame(product_and_prices_dict)
```

```python
# Print out the DataFrame to see the values.

new_product_and_prices
```

```python
# Create a DataFrame using a list of dictionaries, where each dictionary represents a row in the DataFrame with keys 'Product' and 'Price' and values from the two Series you created.

product_and_prices_list = [{'Product': product, 'Price':price} for product, price in zip(products, prices)]

from_list_df = pd.DataFrame(product_and_prices_list)
```

```python
# Print out the DataFrame to see the values.
from_list_df
```
