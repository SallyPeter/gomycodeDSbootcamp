```python
# Begin by importing the necessary libraries, numpy.

import numpy as np
```

```python
# Use the open() function to open csv file and assign the result to a variable.

f = open('Loan_prediction_dataset.csv')

f.readlines()
```

```python
# Use the numpy array to perform some basic statistical analysis on the data, such as finding the mean, median, and standard deviation of the loan amounts.

data = np.genfromtxt('Loan_prediction_dataset.csv', delimiter=',')

data
```

```python
data = np.genfromtxt('Loan_prediction_dataset.csv', delimiter=',', dtype='int64', usecols=8, filling_values=0) # Loading only column 8 as it is the column containing the loan amounts.

data # Note np.genfromtxt converts str to either -1 or your filling value including the column headers.
```

```python
data = np.delete(data, 0) # removing the column header which is the first entry in the array.
data
```

```python
np.size(data)
```

```python
# Use the numpy array to perform some basic statistical analysis on the data, such as finding the mean, median, and standard deviation of the loan amounts.

mean_data = np.mean(data)
print(f'The mean loan amount is {mean_data}')
```

```python
med_data = np.median(data)
print(f'The median loan amount is {med_data}')
```

```python
std_data = np.std(data)
print(f'The stand deviation of the loan amount is {std_data}')
```

```python
max_data = np.max(data)
print(f'The maximum loan amount is {max_data}')
```

```python
min_data = np.min(data)
print(f'The minimum loan amount is {min_data}')
```
