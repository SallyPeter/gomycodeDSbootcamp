### Should you have a missing value, these are likely ways of filling them

```
data['column_name'].fillna(data['column_name'].mode()[0], inplace=True)  # best for numerical columns
data['column_name'].fillna(data['column_name'].mean(), inplace=True)  # best for numerical columns without outliers
data['column_name'].fillna(data['column_name'].mode()[0], inplace=True)  # best for numerical columns with outliers

```



### In a case where each solution above does not reflect real-world data in the perspective of a domain expert ...


```
df.dropna(axis=0, how='any', inplace=True)  # removes rows with NaN
```

````python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
````

````python
baby_names = []
for name in data['Child\'s First Name']:
  baby_names.append(name)

names_str = ''
names_str += " ".join(baby_names)+" "

wordcloud = WordCloud(width = 800, height = 800,
                background_color ='black',
                min_font_size = 10).generate(names_str)
````

````python
# plot the WordCloud image
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
````
